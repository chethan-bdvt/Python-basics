```python
'6.Write a function to demonstrate the behaviour of mutable and immutable arguments'
def demonstrate_mutable_immutable(a, b):
    print("Before modification:")
    print("Immutable (int) a =", a)
    print("Mutable (list) b =", b)

    # Try to modify both
    a += 10         # Creates a new integer object (immutable)
    b.append(100)   # Modifies the same list object (mutable)

    print("\nAfter modification inside function:")
    print("Immutable (int) a =", a)
    print("Mutable (list) b =", b)

# Call the function
x = 5            # Immutable
y = [1, 2, 3]     # Mutable
demonstrate_mutable_immutable(x, y)

print("\nAfter function call:")
print("x =", x)      # Unchanged
print("y =", y)      # Changed
```
