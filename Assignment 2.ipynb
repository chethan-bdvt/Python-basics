```python
  '''
2.Explain the difference between mutable and immutable data types with examples.
-->Mutable Data Types: These are data types whose values can be changed after they are created.
   When you modify a mutable object, you are changing the object in place in memory. Any variables referencing that object will see the changes.
   ex:Lists, Dictionaries, Sets.
   '''
  my_list = [1, 2, 3]
  another_list = my_list
  my_list.append(4)
  print(my_list)
  print(another_list)
  print(id(my_list) == id(another_list))

  '''
  Immutable Data Types: These are data types whose values cannot be changed after they are created.
  Any operation that appears to "modify" an immutable object actually creates a new object with the desired changes, leaving the original object untouched.
  ex:Integers, Floats, Strings, Tuples.
  '''
  my_string = "hello"
  another_string = my_string
  my_string = my_string + " world"
  print(my_string)
  print(another_string)
  print(id(my_string) == id(another_string))
```

```python
'''
17. Describe the time complexity of accessing elements in a dictionary?
-->The time complexity of accessing elements (retrieving a value using its key) in a dictionary is:
    1.Average Case: O(1) - Constant Time
        This is the primary advantage of dictionaries. On average, no matter how large the dictionary is, accessing an element takes roughly the same amount of time. This is because dictionaries use a hash table. The hash function quickly computes an index, allowing direct access to the bucket where the key-value pair is likely stored.

    2.Worst Case: O(N) - Linear Time
        The worst-case scenario occurs in very rare situations, primarily due to hash collisions.
        Python's dictionary might have to perform a linear search through a "chain" of items in a single bucket. In such an extreme case, the time to access an element could depend linearly on the number of elements (N).
  '''
```

```python
'''
18. In what situations are lists preferred over dictionaries?
--> 1.Order Matters:
        When the order of elements is significant and needs to be preserved (e.g., a sequence of steps, historical data, elements needing to be processed in a specific order). Lists maintain insertion order.

    2.Indexing by Position:
        When need to access elements primarily by their numerical position (index) rather than by a meaningful key.

    3.Homogeneous Data Collections (often):
        While lists can hold mixed types, they are often used for collections of similar items where the structure is simpler and a key-value mapping is not necessary.

    4.No Unique Identifiers for Each Item:
        When the items in the collection don't inherently have unique, hashable identifiers that can serve as keys.

    5.Small Fixed-Size Collections (sometimes):
        For very small collections where the overhead of hash table management for a dictionary might be unnecessary.

    6.Duplicates are Allowed/Desired:
        If your collection needs to store multiple instances of the same value. Lists allow duplicates, while dictionaries (keys) and sets do not.

    7.Implementing Stacks or Queues:
        Lists can easily be used to implement basic stack (LIFO) and queue (FIFO) behaviors using append()/pop() and append()/pop(0) respectively.
  '''
```

```python
'''
19. Why are dictionaries considered unordered, and how does that affect data retrieval?
-->Dictionaries in Python were considered unordered because they did not guarantee to preserve the insertion order of elements. The order in which key-value pairs were stored internally (due to hashing) was not necessarily the order in which they were inserted, and this order could even change across different runs of the same program or Python versions.

  Affect on Retrieval:
     The "unordered" nature meant that if iterated through a dictionary, couldn't rely on the order of the items being consistent with their insertion. If needed ordered retrieval, it would have to get the keys, sort them, and then iterate.

'''
```
