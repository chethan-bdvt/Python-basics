```python
"13. How would you check if a file exists before attempting to read it in Python"

class FileExists:
  def __init__(self, file_name):
    self.file_name = file_name

  def read(self):
    try:
      file = open(self.file_name, 'r')
      print(f"File {self.file_name} exists")
    except FileNotFoundError as e:
      print(f"Provide file {self.file_name} doesn't exist")
    except Exception as e:
      print(f"An error occured: {e}")
    else:
      print(f'content of the file is-->\n{file.read()}')
      file.close()

file = "file.txt"
open(file, 'w').write("This is a content written into file for reading\nThis file have multiple lines to read")
obj1 = FileExists(file)
obj1.read()

obj2 = FileExists("file1.txt")
obj2.read()
```

```python
"16.Demonstrate how to use memory profiling to check the memory usage of a small program"

from memory_profiler import profile

class Memory:
  def __init__(self, num1, num2):
    self.num1 = num1
    self.num2 = num2

  @profile
  def memory(self):
    try:
      divide = self.num1/self.num2
      print(divide)
    except ZeroDivisionError as e:
      print(f'An error occured {e}')
      return 1
    except TypeError as e:
      print(f'An error occured {e}')
      return 2
    except Exception as e:
      print(f'General exception {e}')
      return 3

obj1 = Memory(10,0)
obj1.memory()

obj1 = Memory(10, 5)
obj1.memory()
```

```python
"18. How would you implement a basic logging setup that logs to a file with rotation after 1MB"

import logging
from logging.handlers import RotatingFileHandler
import os

class MemoryRotation:
    def __init__(self, file_name):
        self.file_name = file_name
        self.logger = logging.getLogger('RotatingLogger')
        self.logger.setLevel(logging.INFO)

        # Create a rotating file handler
        # maxBytes is 1MB (1024 * 1024 bytes)
        # backupCount is the number of backup files to keep
        handler = RotatingFileHandler(self.file_name, maxBytes=1024 * 1024, backupCount=5)
        formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
        handler.setFormatter(formatter)

        # Add the handler to the logger
        self.logger.addHandler(handler)

    def log_message(self, level, message):
        if level == 'info':
            self.logger.info(message)
        elif level == 'warning':
            self.logger.warning(message)
        elif level == 'error':
            self.logger.error(message)
        elif level == 'critical':
            self.logger.critical(message)
        elif level == 'debug':
            self.logger.debug(message)
        else:
            print(f"Invalid log level: {level}")

# Example usage:
log_file = "rotating_log.log"

# Remove the log file if it exists from previous runs
if os.path.exists(log_file):
    os.remove(log_file)

rotator = MemoryRotation(log_file)

# Log some messages to demonstrate rotation
for i in range(200): # Log enough messages to trigger rotation
    rotator.log_message('info', f"This is log message number {i}")

print(f"Check the file '{log_file}' and potentially '{log_file}.1', '{log_file}.2', etc. for rotated logs.")
```

```python
"20. How would you open a file and read its contents using a context manager in Python"

class contextmanager:
  def __init__(self, file_name):
    self.file_name = file_name

  def read(self):
    try:
      with open(self.file_name, 'r') as f:
        print(f.read())
    except FileNotFoundError:
      print(f"Error: The file '{self.file_name}' was not found.")
    except Exception as e:
      print(f"An error occurred: {e}")

file_name = "my_example_file.txt"

with open(file_name, "w") as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")

file_reader = contextmanager(file_name)
file_reader.read()

file_reader_error = contextmanager("non_existent_file.txt")
file_reader_error.read()
```

```python
"21. Write a Python program that reads a file and prints the number of occurrences of a specific word"

class StringOccurence:
  def __init__(self, file_name):
    self.file_name = file_name

  def findoccurence(self,string):
    try:
      with open(self.file_name, 'r') as f:
        content = f.read()
        print(content.count(string))
    except FileNotFoundError:
      print(f"Error: The file '{self.file_name}' was not found.")
    except Exception as e:
      print(f"An error occurred: {e}")

file = "file.txt"
open(file, 'w').write("This is a content written into file for reading\nThis file have multiple lines to read")
obj1 = StringOccurence(file)
obj1.findoccurence("read")
obj1.findoccurence("write")
obj1.findoccurence("This")
```

```python
"22. How can you check if a file is empty before attempting to read its contents"

class EmptyFile:
  def __init__(self, file_name):
    self.file_name = file_name

  def read(self):
    try:
      with open(self.file_name, 'r') as f:
        content = f.read()
        if(content == ""):
          raise Exception("File is empty")
    except FileNotFoundError:
      print(f"Error: The file '{self.file_name}' was not found.")
    except Exception as e:
      print(f"An error occurred: {e}")

file = "file.txt"
open(file, 'w').write("")
obj1 = EmptyFile(file)
obj1.read()
```

```python
"23. Write a Python program that writes to a log file when an error occurs during file handling."

import logging
class LogWhenError:
  def __init__(self, file_name):
    self.file_name = file_name
    logging.basicConfig(filename=self.file_name,format='%(asctime)s - %(levelname)s - %(message)s',level="INFO""WARN""ERROR")

  def write(self,string):
    try:
      with open(self.file_name, 'w') as f:
        if(string == ""):
          logging.warn("Warning: The string is empty.")
        else:
          f.write(string)
    except FileNotFoundError:
      logging.error(f"Error: The file '{self.file_name}' was not found.")
    except Exception as e:
      logging.error(f"An error occurred: {e}")

file = "file.txt"
open(file, 'w').write("This is a content written into file for reading\nThis file have multiple lines to read")
obj1 = LogWhenError(file)
obj1.write("This is a new line")
obj1.write("")

obj2 = LogWhenError("")
obj2.write("This is a content written into file for reading\nThis file have multiple lines to read")
```
