```python
'''
6. Explain the concept of generators in Python and how they are defined.
--> Generators are a special type of iterable in Python that allow you to create iterators in a memory-efficient way. Instead of building and storing an entire sequence in memory (like a list), generators produce items one at a time, only when requested. This makes them ideal for processing large datasets or infinite sequences.

    How they are defined:
      Generators are defined like regular functions, but instead of using the return statement to send back a final result, they use the yield statement to produce a sequence of values. Any function that contains at least one yield statement automatically becomes a generator function. It returns a generator object (an iterator), not the actual values.
'''
```

```python
'''
7. What are the advantages of using generators over regular functions?
-->Advantages of Generators over Regular Functions
      a.Memory Efficiency (Lazy Evaluation):
          This is the primary advantage. Generators produce items one by one, on demand. They don't store the entire sequence in memory, which is crucial when dealing with very large or infinite sequences. Regular functions that return lists would load everything into memory at once, potentially causing MemoryError for large datasets.

      b.Performance Improvement:
          Because they generate values lazily, generators can start processing data before the entire sequence is computed, leading to better perceived performance for large operations.

      c.Readability and Simplicity:
          Generator functions often look like regular functions, making them easy to write and understand, especially for complex iteration patterns. The yield keyword simplifies the creation of custom iterators.

      d.Pipelining:
          Generators are excellent for building data processing pipelines. You can chain multiple generator functions together, where the output of one generator feeds into the input of another, creating a highly efficient and readable data flow.

      e.State Retention:
          Generators automatically handle the saving and restoring of their internal state between yield calls, eliminating the need for explicit state management in your code.
'''
```

```python
'''
8. What is a lambda function in Python and when is it typically used?
--> Lambda functions are anonymous, small, and single-expression functions.

    Syntax: lambda arguments: expression
        arguments: These are the inputs to the lambda function, similar to parameters in a regular function.
        expression: This is a single expression whose result is implicitly returned by the lambda function.

    Characteristics:
        Anonymous:
          They don't have a name.

        Single Expression:
          They can only contain one expression.

        Implicit Return:
          The result of the expression is automatically returned; you don't use the return keyword.

        Concise:
          They are designed for simple, short operations.

    When is it typically used?
      a.As arguments to map(), filter(), sorted(), min(), max():
          These built-in functions often require a function to specify how elements should be transformed, filtered, or compared. Lambdas provide a concise way to define these inline.

      b.For simple callbacks:
          GUI programming or event handling, lambdas can be used for simple callback functions.

      c.Key functions for sorting or grouping:
          When you need a custom key for sorting lists of complex objects (e.g., sorting a list of dictionaries by a specific key), lambdas are very handy.
'''

# 1. Basic lambda function
add_two_numbers = lambda x, y: x + y
print(f"Sum using lambda: {add_two_numbers(5, 3)}")

# 2. Using with map()
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x * x, numbers))
print(f"Squared numbers (using map): {squared_numbers}")

# 3. Using with filter()
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers (using filter): {even_numbers}")

# 4. Using with sorted()
students = [
    {'name': 'Alice', 'age': 25},
    {'name': 'Bob', 'age': 20},
    {'name': 'Charlie', 'age': 30}
]

# Sort students by age using a lambda as the key
sorted_students = sorted(students, key=lambda student: student['age'])
print(f"Sorted students by age: {sorted_students}")

# 5. Using with min()/max()
data_points = [(1, 5), (3, 2), (2, 8)]
# Find the point with the maximum y-coordinate
max_y_point = max(data_points, key=lambda point: point[1])
print(f"Point with max Y: {max_y_point}")
```

```python
'''
9. Explain the purpose and usage of the `map()` function in Python.
--> The map() function in Python is used to apply a given function to each item in an iterable (like a list or tuple) and return a new iterator that yields the results.

      Syntax: map(function, iterable)
        function: The function to which each item of the iterable will be passed. This can be a built-in function, a user-defined function, or a lambda function.
        iterable: One or more iterable objects (e.g., lists, tuples, strings) whose elements will be passed to the function. If multiple iterables are provided, the function must accept that many arguments, and map() will stop when the shortest iterable is exhausted.
'''
```

```python
'''
10. What is the difference between map(), reduce(), and filter() functions in Python?
--> a.map():
        1.Transformation: Applies a function to each item to transform it into a new item.
        2.syntax: map(function, iterable)
        3.Return Type: Returns an iterator with the transformed items.
        4.Arguements: Takes a function and one or more iterables as arguments.
        5.Example: map(lambda x: x * 2, [1, 2, 3, 4])

    b.reduce():
        1.Accumulation: Applies a function to the first two items in the iterable, then applies the same function to the result and the next item, and so on.
        2.syntax: reduce(function, iterable)
        3.Return Type: Returns a single value.
        4.Arguements: Takes a function and
        5.Example: reduce(lambda x, y: x + y, [1, 2, 3, 4])

    c.filter():
        1.Filtering: Filters out items from an iterable based on a specified condition.
        2.syntax: filter(function, iterable)
        3.Return Type: Returns an iterator with the filtered items.
        4.Arguements: Takes a function that returns a boolean value and an iterable.
        5.Example: filter(lambda x: x % 2 == 0, [1, 2, 3, 4])

    Key Differences:
        map(): Transforms each element independently. "One in, one transformed out." (though technically, it's one or more iterables in, one transformed iterable out)
        filter(): Selects elements based on a condition. "Many in, some out."
        reduce(): Combines elements sequentially to produce a single result. "Many in, one out."
'''
```

```python
"5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms."
class fibonacci_sequence:
  def __init__(self,n):
    self.n = n
    self.a = 0
    self.b = 1
    self.count = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.count < self.n:
      result = self.a
      self.a,self.b = self.b,self.a+self.b
      self.count+=1
      return result
    else:
      raise StopIteration

fib = fibonacci_sequence(10) # Create an instance of the class
for i in fib:
  print(i)
```

```python
"6. Write a generator function in Python that yields the powers of 2 up to a given exponent."

def power_of_two(n):
  power = 0
  while power <= n:
    yield 2**power
    power+=1
Pow = power_of_two(3)
print(next(Pow))
print(next(Pow))
print(next(Pow))
```

```python
"7. Implement a generator function that reads a file line by line and yields each line as a string."
def read_file(filepath):
  try:
    with open(filepath, 'r') as f:
      for line in f:
        yield line
  except FileNotFoundError:
    print(f"File not found: {filepath}")
  except Exception as e:
    print(f"An error occurred: {e}")

def write_file(filepath, content):
  try:
    with open(filepath, 'w') as f:
      f.write(content.strip())
  except Exception as e:
    print(f"An error occurred: {e}")

file = "sample.txt"
file_content = '''
This is a sample file.
Created for practicing generators.
Which is a PWSkills Assignment.
'''

write_file(file, file_content)
line = read_file(file)
while(next(line)):
  print(next(line)

```
